```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
sigma = lambda x, y: math.exp(-(x**2 + y**2))
```

```python
def approx_reimann_sum_2D(f, a, b, arg_const, h):
    sum = 0
    N = int((b-a)/h) + 1
    y_arr = np.linspace(a, b, N)
    funcvec = np.vectorize(f, otypes=[float])
    x_arr = np.ones(N)*arg_const
    f_arr = funcvec(x_arr, y_arr)
    #print(x_arr, y_arr, f_arr)
    plt.plot(x_arr, y_arr)
    return np.sum(f_arr)*h
```

```python
def integral_2D(f, a, b):
    N = 2e3
    xarr = np.linspace(a, b, int(N+1))
    dx = 1./N
    dy = 1./N
    I_tilde = 0

    limarr = []

    for x in xarr:
        y_lim = abs(math.sqrt(1 - x**2))
        limarr.append(y_lim)
        I_tilde += dx*approx_reimann_sum_2D(f, -y_lim, y_lim, x, dy)

    plt.show()
    print(I_tilde)
```

```python
integral_2D(sigma, -1, 1)
```

```python
integral_2D(sigma, 0, 1)
```

```python
integral_2D(sigma, -1, 0)
```

It is bizzare how integrals from -1 to 0 and 0 to 1 evaluate to 0.99..., but the integral from -1 to 1 doesn't evaluate to 1.98...
