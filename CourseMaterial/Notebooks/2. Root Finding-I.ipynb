# Root Finding
Finding values where a function is zero. A common example is solving dispersion relation.
$$f(\omega, k) = 0$$
$k = \frac{2\pi}{\lambda}$ and $\omega = \frac{2\pi}{T}$. A simple relation in case of light waves is: 
$$\omega = ck$$
Given $k$, find $\omega$.
$$\omega - kc = 0 \\ k = \omega/c$$
A polynomial of order 3 is the max that can be solved analytically (the formula like the quadratic formula doesnt exist for higher degree). \\
Another problem is for solving for the wave function $\psi$ in a potential well. Analytical solutions require finding roots of some equation. \\
Another problem is finding the trajectory of planets...? Finding apogee and perigee.

### Bisection Method
If we know two points $x_0$ and $x_1$ such that $f(x_0) < 0$ and $f(x_1) > 0$, then we can use the Bisection method to find a root of $f(x)$. If $f(x_0)f(x_1) < 0$ then there exists a point $ x_0 < x < x_1$ such that $f(x) = 0$.

After getting the interval $[x_0, x_1]$, we can find the midpoint $x_2 = \frac{x_0 + x_1}{2}$. 

If $f(x_2) = 0$ then we have found the root.

If $f(x_2 ) < 0$ then we replace $x_0$ with $x_2$ and 

if $f(x_2) > 0$ then we replace $x_1$ with $x_2$. 

We repeat this process until we find the root or until the interval $[x_0, x_2]$ is small enough.
This is just a binary search.


We need to define a threshold. If the function value goes below that threshold we take it to be 0. Let threshold `eps`. Our exit condition is if $f(x) < \,$`eps`

```python
eps = 1e-10

f = lambda x: math.tan(x) - math.exp(x)

def root(f, a, b):
    #a, b = 0, 0
    while True:
        c = (a+b)/2.
        if abs(f(c)) < eps:
            return c
        elif f(a)*f(c) < 0:
            b = c
        else:
            a = c
```
