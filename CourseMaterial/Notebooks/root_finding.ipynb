# Finding root
## Bisection method

```python
import math 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
def func(x):
    return np.tan(x) - np.exp(x)
```

```python

eps = 1E-10

error = 1
a = 0
b = np.pi/2.0
while eps < error:
    x0 = (a+b)/2
    if func(x0)*func(a) < 0:
        b = x0
    else:
        a = x0
    error = np.abs(func(x0))

print(x0)
```

```python
x = np.linspace(0, np.pi/2 - 0.01, 500)
y = func(x)
  
plt.plot(x,y)
plt.grid()
plt.xlabel("x values")
plt.ylabel("function")
plt.title("graaph")
plt.scatter(x0,func(x0), c="r")
plt.show()
```

```python
def func(x):
    return np.tan(x) - np.exp(x)

def derivative(x):
    h = .00000001
    return (func(x+h) - func(x-h))/(2*h)
```

```python
x0 = 1.5
error = 1
eps = 1e-10

while(error>eps):
    f_val = func(x0)
    f_prime = derivative(x0)
    x1 = x0 - (f_val/f_prime)
    error = func(x1)
    print("x0 = ",x0, "x0 = ", x1)

    x0 = x1

print(x0)
```
