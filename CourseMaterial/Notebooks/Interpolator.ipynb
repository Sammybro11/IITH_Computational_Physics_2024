```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 11)
f_k = np.sin(x)
```

```python
y = np.linspace(0, 10, 1000)

def langrange(x, f_k, y):
    """
    Lagrange interpolation polynomial.
    """
    n = len(f_k)
    p = np.zeros_like(y)
    for i in range(n):
        L = np.ones_like(y)
        for j in range(n):
            if i != j:
                L *= (y - x[j]) / (x[i] - x[j])
        p += f_k[i] * L
    return p

ans = langrange(x, f_k, y)
```

```python
plt.plot(y, ans)
plt.scatter(x, f_k, color='red')
plt.show()
```

```python
lag_pols = []

def lagrange_interpolation(x_points, y_points, x_values):
    """
    General implementation of Lagrange interpolation.

    Parameters:
    x_points (numpy.ndarray): Known x-coordinates.
    y_points (numpy.ndarray): Known y-coordinates (function values at x_points).
    x_values (numpy.ndarray): Points at which to evaluate the interpolation.

    Returns:
    numpy.ndarray: Interpolated values at x_values.
    """
    n = len(x_points)
    interpolated_values = np.zeros_like(x_values)
    for i in range(n):
        L = np.ones_like(x_values)
        for j in range(n):
            if i != j:
                L *= (x_values - x_points[j]) / (x_points[i] - x_points[j])
        interpolated_values += y_points[i] * L
        lag_pols.append(L*y_points[i])
    return interpolated_values

# Example usage:
interpolated_result = lagrange_interpolation(x, f_k, y)
#interpolated_result
```

```python
plt.plot(y, interpolated_result)
plt.scatter(x, f_k, color='red')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
for i in range(len(lag_pols)):
    plt.plot(y, lag_pols[i], label=f"Lagrange Polynomial {i+1}")
plt.legend()
plt.scatter(x, f_k, color='red')
plt.title(f"Lagrange Polynomial {i+1}")
plt.show()
```
