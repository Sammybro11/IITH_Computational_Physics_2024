```python
import numpy as np
import matplotlib.pyplot as plt

N = 1000

X = np.random. rand(N)
Y = np.tan(np.pi*X/2)

x_space = np.linspace(0, 1, N)

plt.hist(Y, bins=100, density=True, alpha=0.5, color='g')
plt.title('Histogram of Y = tan(pi*X/2)')
plt.plot(1/(1+x_space**2), x_space, 'r-', lw=2, label='PDF of Cauchy distribution')
plt.show()
```
