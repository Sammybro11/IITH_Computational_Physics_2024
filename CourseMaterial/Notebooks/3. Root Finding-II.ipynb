## Newton Ralphson Method

```python
import math
import numpy as np
```

```python
eps = 1e-10

def new_ralph(f, f_dash, x_0):
    err = f(x_0)
    while err > eps:
        x1 = x_0 - f(x_0)/f_dash(x_0)
        err = abs(f(x1) - f(x_0))
        print(err)
        x_0 = x1
    return x_0
```

```python
x0 = math.pi/2 - 0.01

f = lambda x: math.tan(x) - math.exp(x)
f_dash = lambda x: math.cos(x)**-2 - math.exp(x)

new_ralph(f, f_dash, x0)
```

It is not guranteed that we get the solution in this method. It depends on the starting point x_0. We might not get the root we desired because of the local minima or maxima. We might get stuck in a local minima or maxima, where the derivative is zero and the step blows up.
